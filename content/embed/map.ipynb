```python
import pandas as pd
import folium
from IPython.display import display #Display HTML inline

df = pd.read_excel('nu-police-data-may2016.xlsx')

def create_clustered_map(year, df):
    
    #Groups crime incidents by year
    df = df.groupby(df['Date & Time: Occurred'].map(lambda x: x.year))
    df = df.get_group(year)
    
    display(df.head())
    
    #Creates a list of tuples with (latitude, longitude, incident type)
    coordinates = zip((x for x in df.ix[:,8]), (y for y in df.ix[:,9]), (z for z in df.ix[:,6]))

    evanston = [42.050803654475, -87.6760827563392]
    
    clustermap = folium.Map(location=evanston, zoom_start=14)
    marker_cluster = folium.MarkerCluster().add_to(clustermap)
    
    #Adds each marker with the incident type as a label
    for coordinate in coordinates:
        folium.Marker([coordinate[0], coordinate[1]],
              popup=coordinate[2],
              icon = folium.Icon(color='red', icon='ok-sign'),
              ).add_to(marker_cluster)
    display(clustermap)
    clustermap.save(str(year)+'clustermap.html')
    
create_clustered_map(2016, df)
```
